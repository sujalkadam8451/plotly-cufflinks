# Visualizing statistical relationships

Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because, when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship.

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

```python
tips['smoker'].value_counts()
```

# Hue
We can add another dimension in our plot with the help of hue as it gives color to the points and each color has some meaning attached to it.

The hue semantic is categorical. Thatâ€™s why it has a different color palette. If the hue semantic is numeric, then the coloring becomes sequential.

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size')
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips,hue = 'smoker', 
            style = 'time', size = 'size')
plt.show()
```

# Line Plot

```python
from numpy.random import randn
```

```python
df = pd.DataFrame(dict(time = np.arange(500),value = randn(500).cumsum())) #Random Data Genrated
```

```python
df.head()
```

```python
g=sns.relplot(x = 'time', y = 'value', kind = 'line', data = df)
g.fig.autofmt_xdate(bottom=0.1, rotation=60)
plt.show()
#The autofmt_xdate() method figure module of matplotlib library is used to 
#rotate them and right align them.
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
plt.show()
```

```python
# The seaborn ci code simply computes the percentile limits. 
#The actual mean, the standard deviation, etc. will appear in the percentiles routine
#Size of confidence intervals to draw around estimated values.
#If â€œsdâ€,  draw the standard deviation of the observations.
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = 'sd') 
plt.show()
```

```python
sns.lineplot(x = 'timepoint', y  = 'signal', style = 'event', hue = 'region', 
             data = fmri, markers = True, ci = 68, err_style='bars')#band
plt.show()
```
