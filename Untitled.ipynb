=>Seaborn is a data visualization library built on top of Matplotlib.
=>It provides a high-level interface for drawing attractive and informative statistical graphics.
=>It integrates closely with pandas DataFrames.

# Key Features
=>Built on top of Matplotlib (simpler and more aesthetic syntax)
=>Integrates well with Pandas DataFrames
=>Supports complex visualizations with minimal code
=>Comes with built-in themes, color palettes, and sample datasets
=>Supports statistical visualization (regression, distributions, etc.)
=>Automatically handles DataFrame column names and categories

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load built in data set
df= sns.load_dataset('tips')
#display few records
print(df.head())
```

```python
#Used to show relationships between two variables.
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
#Used for time series or trends.
sns.lineplot(x="size", y="tip", data=df)
plt.show()
```

# Bar Plot:
=>A bar plot displays the aggregate (summary) value of a numerical variable for each category of a categorical variable.
=>It's mainly used to compare averages, sums, or other aggregates across categories.


```python
sns.barplot(x="day", y="total_bill", data=df)
plt.show()
```

# Box Plot:
=>A Box Plot (also called a Box-and-Whisker Plot) is a statistical chart used to visualize
the distribution of a numerical variable

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

A violin plot is a combination of a box plot and a kernel density plot.
It shows:

=>The distribution of numerical data across categories
=>The probability density (using a smooth curve)
=>The median and quartiles (like a box plot)

```python
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.countplot(x="day", data=df)
plt.show()
```

```python
sns.histplot(df["total_bill"], bins=10, kde=True)
plt.show()
```

```python
sns.kdeplot(df["tip"])
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="sex", data=df)
plt.show()
```

```python
import numpy as np

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df, hue="sex")
plt.show()
```

```python
sns.set_style("darkgrid")   # Options: white, dark, whitegrid, darkgrid, ticks
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
sns.set_palette("husl")
sns.boxplot(x="day", y="tip", data=df)
plt.show()
```

```python
g = sns.FacetGrid(df, col="sex")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```
